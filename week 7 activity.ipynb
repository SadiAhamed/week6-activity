# 2.Examine the data using visualisation

```python
df=pd.DataFrame(data)
sa=df[["country","Surface area (km2)"]]
sa5=sa.sort_values(by="Surface area (km2)",ascending=False).head(5)
sa5
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="country",y="Surface area (km2)",data=sa5)
plt.title("top 5 countries with highest surface area")
plt.show()
```

```python
ds=df.sort_values(by="Population in thousands (2017)",ascending=True).head(5)
plt.figure(figsize=(10,8))
sns.barplot(x="country",y="Population in thousands (2017)",data=ds)
plt.title("bottom 5 countries with lowest population")
plt.show()
```

```python
dx=df.sort_values(by="GDP per capita (current US$)",ascending=False).head(7)
plt.figure(figsize=(10,8))
sns.barplot(x="country",y="GDP per capita (current US$)",data=dx)
plt.title("top 7 countries with highest GDP")
plt.show()
```

# 5.Find out which region has more consistent with respect to GDP per capita?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats
data=pd.read_excel(r'D:\activity\Dataset_w6.xlsx')
data.head()
```

```python
df=pd.DataFrame(data)
data2=df.groupby("Region")["GDP per capita (current US$)"].describe()
data2
```

```python
dfn=pd.DataFrame(data2)
dfn["cv"]=(data2["std"]/data2["mean"])*100
dfn.sort_values(by="cv",ascending=True)
```

### the region oceania is most consistent since its coefficient of variation is minimum
